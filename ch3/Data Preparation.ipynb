# Data Preparation

We have a .xls file containing financial information and our goal is to prepare the data for training.

```python
import pandas as pd
data = pd.read_excel("default of credit card clients.xls", skiprows=1) #skiprows removes the first row, which is just a second row of headers
data.head()
```

```python
print("Rows:", data.shape[0], ", columns:", data.shape[1])
```

```python
data_clean = data.drop(columns=["ID","SEX"])
data_clean.head()
```

Now we have removed irrelevant fields from the data, we need to check for missing values. If we are missing values, we'll need to know what percentage of each columns values are missing

```python
total = data_clean.isnull().sum()
percent = (data_clean.isnull().sum()/data_clean.isnull().count()*100)
pd.concat([total, percent], axis=1,
          keys=['Total', 'Percent']).transpose()
```

```python
outliers = {}
for i in range(data_clean.shape[1]):
    min_t = data_clean[data_clean.columns[i]].mean() - (3 * data_clean[data_clean.columns[i]].std())
    max_t = data_clean[data_clean.columns[i]].mean() + (3 * data_clean[data_clean.columns[i]].std())
    count = 0
    for j in data_clean[data_clean.columns[i]]:
        if j < min_t or j > max_t:
            count+=1
    percentage = count/data_clean.shape[0]
    outliers[data_clean.columns[i]] = "%.3f" % percentage
print(outliers)
```

There are very few outliers that will affect this data so we can move on.

Now we need to check for class imbalance. If this data set contains 90% of records saying they defaulted on the following month, then that would lead to bias.

```python
target = data_clean["default payment next month"]
yes = target[target == 1].count()
no = target[target == 0].count()
print("yes: %: " + str(yes/len(target)*100) + " - no %: " + str(no/len(target)*100))
```

```python
# the above results can also be displayed in matplotlib
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(10,5))
plt.bar("yes", yes)
plt.bar("no", no)
ax.set_yticks([yes,no])
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

Since we have significantly more "No" than "Yes", we'll have to even out the data a bit

```python
data_yes = data_clean[data_clean["default payment next month"] == 1]
data_no = data_clean[data_clean["default payment next month"] == 0]
over_sampling = data_yes.sample(no, replace=True, random_state=0)
data_resampled = pd.concat([data_no, over_sampling], axis=0)

# now that we've resampled the data, we split the features from the target
data_resampled = data_resampled.reset_index(drop=True)
x = data_resampled.drop(columns=["default payment next month"])
y = data_resampled ["default payment next month"]
```

Finally we rescale the values of the features matrix in order to avoid introducing bias in to the model

```python
x = (x-x.min())/(x.max()-x.min())
x.head()
```

```python
final_data = pd.concat([x,y], axis=1)
final_data.to_csv("dccc_prepared.csv", index=False)
```
